```python
import csv
import numpy as np
import numpy.linalg as ln
import pylab as pl
import matplotlib.pyplot as plt
import copy
#import PFFileIO as pfio
%matplotlib inline
```

```python
plt_colors=['r','g','b','m','c']
plt_labels=['1','2','3','4','5']
```

```python
results_filenames = ['rs/meta_res_004/results_rep0001_mop1_mft1_mul1.33_mhf005_mhm50.txt',
                     'rs/meta_res_004/results_rep0001_mop1_mft1_mul1.33_mhf125_mhm50.txt']
gt_filename = 'ds/gt_fps30.txt'
results = [];
for f in results_filenames:
    with open(f) as inputfile:
        results.append(list(csv.reader(inputfile)))

with open(gt_filename) as inputfile:
    gt = list(csv.reader(inputfile))
```

```python
kinematic_res = copy.deepcopy(results)
for f in kinematic_res:
    for i in f:
        del i[27:len(i)]
        
for f in results:
    for i in f:
        del i[0:27]
        
```

```python
#Plot Meta
ndims=len(results[0][0])
meta_gt = np.ones(ndims,dtype=np.float)
#x_vals=
diff = []
for idx,f in enumerate(results):
    diff = []
    for i in f:
        inp = np.float_(i)
        n1 = np.linalg.norm(inp-meta_gt)
        #print n1, inp
        diff.append(n1)
        #print diff
        plt.plot(diff,color=plt_colors[idx],label=plt_labels[idx])
#plt.legend()
plt.ylabel('E(meta)')
plt.show()
    
#plot kinematic
ndims=len(kinematic_res[0][0])
#x_vals=
diff = []
for idx,res_files in enumerate(kinematic_res):
    diff = []
    for fidx,i in enumerate(res_files):
        inp = np.float_(i)
        gt_np = np.float_(gt[fidx])
        n1 = np.linalg.norm(inp-gt_np)
        #print n1, inp
        diff.append(n1)
        #print diff
        plt.plot(diff,color=plt_colors[idx],label=plt_labels[idx])
#plt.legend()
plt.ylabel('E(kinematic)')
plt.show()
```
