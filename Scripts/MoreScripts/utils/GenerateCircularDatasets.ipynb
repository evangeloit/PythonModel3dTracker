```python
import numpy as np
```

```python
input_gt_path = 'ds/gt_fps30.txt'
gt_in = np.loadtxt(input_gt_path, dtype=float, comments='#', delimiter=' , ')  
```

```python
rgt_in = gt_in[::-1]
gt_double = np.concatenate((gt_in, rgt_in))
```

```python
start_frames = np.arange(50,651,50)
filename_templ = 'ds/gtcirc_sta%03d_fps30.txt'
l = len(gt_double)
for f in start_frames:    
    cur_filename = filename_templ % f
    gt_cur = np.concatenate((gt_double[f:l],gt_double[0:f]))
    print cur_filename, len(gt_cur), gt_cur[0]
    np.savetxt(cur_filename,gt_cur,delimiter=' , ')
```
