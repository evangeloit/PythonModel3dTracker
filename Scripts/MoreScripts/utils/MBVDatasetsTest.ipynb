```python
import PyMBVCore as Core
import PyMBVAcquisition as acq
import PyPFUtilsLib as pf
```

```python
ds_folder = "media/pf/bt_datasets_new.xml"
model_name = 'human_ext'
```

```python
ds = pf.HTDataset(ds_folder,model_name)
ds.PrintDatasets()
ds_ids = ds.getDatasetIDs()
print 'Available Datasets:',ds_ids
```

```python
for d in ds_ids:
    ds.genFilenames(d)
    print '---------------Dataset',d
    print 'Streams:',ds.getStreamFilename()
    print 'Calib:',ds.getCalibFilename()
    print 'Frames:',ds.getStartFrame(), ds.getStopFrame()
    meta_info = ds.getMetaDatasetInfo('default',pf.MetaType.MTResParamvec)
    if (meta_info):
        print 'MetaInfo:', meta_info.name, meta_info.type
        print 'MetaInfoRes:', meta_info.getAvailableModels()
        for m in meta_info.getAvailableModels():
            for f in meta_info.getAvailableFrames(m):
                print 'model:',m,'frame:',f,'state:',meta_info.getResult(m,f)
    
```
