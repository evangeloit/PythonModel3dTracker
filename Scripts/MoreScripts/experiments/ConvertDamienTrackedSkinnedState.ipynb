```python
import PyMBVCore as core
import PyMBVDecoding as dec
import PyMBVAcquisition as acq
import PyMBVRendering
import PyMBVLibraries as lib
import PyMBVParticleFilter as pf
import PythonModelTracker.PFHelpers.ModelTrackingResults as mtr
import csv
import PythonModelTracker.Paths as paths
import BlenderMBVLib.AngleTransformations as at
```

```python
did = 'skl_s01_a04_r01'
damien_fname = paths.datasets + '/human_tracking/synthetic/smhad/{0}/hyps.txt'.format(did)
output_json_fname = paths.datasets + '/human_tracking/synthetic/smhad/{0}_gt.json'.format(did)
target_model_name = 'mh_body_male_meta_glbscl'
```

```python
# Load Damien's txt file with states.

states = []
with open(damien_fname) as csvfile:
    reader = csv.reader(csvfile, delimiter=' ')
    for i, row in enumerate(reader):
        if i == 0: dim_names_damien = row
        elif i>=35: states.append([float(s) for s in row])

res = mtr.ModelTrackingResults(did)
for i, s in enumerate(states):
    res.add(i,'damien',s)
dim_names_damien.append(None)
            
#res = mtr.ModelTrackingResults()
#res.load(paths.datasets + '/human_tracking/synthetic/smhad/hyps.json')
```
